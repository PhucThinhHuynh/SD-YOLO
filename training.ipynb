```python
from ultralytics import YOLO

# Load model từ file .pt đã tải
model = YOLO("/kaggle/working/SD-YOLO/runs/detect/train/weights/best.pt")
```

```python
# 4. Chạy test trên VisDrone.yaml
results = model.val(data="VisDrone.yaml", imgsz=640, batch=1, device=0)

# 5. Lấy log tốc độ từ results.speed (dict: preprocess, inference, postprocess, đơn vị ms/img)
speed = results.speed
inference_time = speed['inference']  # ms per image
fps = 1000.0 / inference_time

print(f"Preprocess: {speed['preprocess']:.2f} ms/img")
print(f"Inference: {speed['inference']:.2f} ms/img")
print(f"Postprocess: {speed['postprocess']:.2f} ms/img")
print(f"==> FPS, batch=1: {fps:.2f}")
```
